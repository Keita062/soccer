```python
# 各データにカテゴリ列を追加する例
data1['League'] = 'Premier League'
data2['League'] = 'Championship'
data3['League'] = 'League One'
data4['League'] = 'League Two'
data5['League'] = 'National League'
```

```python
# 各データをリストにまとめる
datasets = [data1, data2, data3, data4, data5]

# 出力用のループ
for i, data in enumerate(datasets, 1):
    print(f"--- Data {i} の 'Round' ユニーク値 ---")
    print(data["Round"].unique(), "\n")
```

```python
# 結合(さまざまなデータを)
data = pd.concat([data1, data2, data3, data4, data5], ignore_index=True)
data
```

```python
# 欠損状況を確認
print(data['FT'].isnull().sum())
print(data['FT'].isnull().mean())  # 欠損割合を確認

# 欠損がある行を表示
print(data[data['FT'].isnull()])
```

２部リーグ以降が欠損のため一時放置

```python
data.columns
```

```python
data['FT'] = data['FT'].str.replace('–', '-', regex=False)

# 'FT' 列に欠損値がない行だけ分割し、Team1scoreとTeam2scoreに格納
data[['Team1score', 'Team2score']] = data['FT'].dropna().str.split('-', expand=True)

# スコア列を float 型に変換（欠損値のまま扱うため）
data['Team1score'] = pd.to_numeric(data['Team1score'], errors='coerce')
data['Team2score'] = pd.to_numeric(data['Team2score'], errors='coerce')

# 結果の確認
data
```

```python
data.columns
```

```python
premier_league_goals = data[data['League'] == 'Premier League'][['Team1score', 'Team2score']].sum().sum()

# 結果の表示
print(f"Premier Leagueのチームでの総得点数: {premier_league_goals}")
```

```python
# "Premier League"の試合に登場したチームの得点を集計
# 1. Team 1 の得点を集計
team1_scores = data[data['League'] == 'Premier League'].groupby('Team 1')['Team1score'].sum()

# 2. Team 2 の得点を集計
team2_scores = data[data['League'] == 'Premier League'].groupby('Team 2')['Team2score'].sum()

# 3. チームごとの得点を合算
total_scores = team1_scores.add(team2_scores, fill_value=0).astype(int).sort_values(ascending=False)

# 結果の表示
print("Premier Leagueのチームごとの得点数:")
print(total_scores)
```
